```python
# Data import
import pandas as pd

# Load CSV file
df = pd.read_csv("BankChurners.csv", skipinitialspace=True)

# Drop two columns
df.drop(
    [
        "CLIENTNUM",
        "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1",
        "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2",
    ],
    axis=1,
    inplace=True,
)

# Make target value binary
df = df.replace(
    {
        "Attrition_Flag": {
            "Existing Customer": 0,
            "Attrited Customer": 1,
        }
    }
)

# Feature selection for modeling
X = df[
    [
        "Total_Trans_Ct",
        "Total_Trans_Amt",
        "Total_Revolving_Bal",
        "Total_Amt_Chng_Q4_Q1",
        "Total_Ct_Chng_Q4_Q1",
        "Total_Relationship_Count",
        "Months_Inactive_12_mon",
        "Contacts_Count_12_mon",
        "Customer_Age",
        "Credit_Limit",
        "Months_on_book",
    ]
]

# Selecting the "Attrition_Flag" column as the target
y = df["Attrition_Flag"]

X
```

```python
# eXtreme Gradient Boost Classifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report,
)
import matplotlib.pyplot as plt
import xgboost as xgb

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=12
)

# Initializing the xgboost model
xgb = xgb.XGBClassifier(
    scale_pos_weight=5.25,
    learning_rate=0.05,
    max_depth=4,
    min_child_weight=1,
    n_estimators=500,
    subsample=0.4331229815476772,
    colsample_bytree=0.5242723167060505,
    reg_lambda=7.049263425549493e-06,
    reg_alpha=0.015123845878697514,
)

# Fitting the model to the training data
xgb.fit(X_train, y_train)

# Fitting for the tree plotting
plot = xgb.fit(X_train, y_train)

# Saving the model in binary format
xgb.save_model("./model/xgbmodel.bin")

# Making predictions on the test set
y_pred = xgb.predict(X_test)

# Creating, printing and saving the confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=xgb.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=xgb.classes_)
disp.plot()
plt.savefig("./visuals/confusion_matrix.png", bbox_inches="tight")
plt.show()

# Printing the classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Model tree plot with matplotlib
from xgboost import plot_tree
import matplotlib.pyplot as plt

# Plotting model tree and save png
plot_tree(plot, num_trees=1, rankdir="LR")
plt.gcf().set_dpi(800)
plt.savefig("./visuals/xgb_matplotlib_tree.png", dpi=800, bbox_inches="tight")
plt.show()
```

```python
# Model tree plot with dtreeviz
import dtreeviz

# Visualizing the decision tree
viz = dtreeviz.model(
    xgb,
    X_train,
    y_train,
    target_name="Attrition_Flag",
    feature_names=X.columns.tolist(),
    class_names=["0", "1"],
    tree_index=1,
)

viz.view(fancy=True, orientation="UD")
```
